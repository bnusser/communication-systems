Convolution is a mathematical operation that combines two sequences (or functions) to produce a third sequence expressing how the shape of one is modified by the other. In discrete time, the convolution of sequences  $x[n]$ and $h[n]$ is defined as:

$$
[ y[n] = (x * h)[n] = \sum_{k=-\infty}^{\infty} x[k] \cdot h[n - k] ]
$$

It is widely used in signal processing to analyze the effect of a system (with impulse response $h[n]$) on an input signal $x[n]$.

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array([3, 4, 6, 1, 5])
h1 = np.array([1, 2, 3, 4, 5])
h2 = np.array([5, 2, 5, 1, 5])

# Compute convolutions
a = np.convolve(x, h1)
b = np.convolve(a, h2)
a1 = np.convolve(x, h2)
b1 = np.convolve(a, h1)
a2 = np.convolve(h1, h2)
b2 = np.convolve(x, a2)


fig, axes = plt.subplots(3, 3, figsize=(15, 12))

# First row
axes[0, 0].stem(x)
axes[0, 0].set_title('x[n]')
axes[0, 1].stem(h1)
axes[0, 1].set_title('h1[n]')
axes[0, 2].stem(h2)
axes[0, 2].set_title('h2[n]')

# Second row
axes[1, 0].stem(a)
axes[1, 0].set_title('x[n] * h1[n]')
axes[1, 1].stem(b)
axes[1, 1].set_title('x[n] * h1[n] * h2[n]')
axes[1, 2].stem(a1)
axes[1, 2].set_title('x[n] * h1[n]')

# Third row
axes[2, 0].stem(b1)
axes[2, 0].set_title('x[n] * h2[n] * h1[n]')
axes[2, 1].stem(a2)
axes[2, 1].set_title('h1[n] * h2[n]')
axes[2, 2].stem(b2)
axes[2, 2].set_title('h1[n] * h1[n] * h2[n]')

# Formatting
for ax_row in axes:
    for ax in ax_row:
        ax.grid(True)
        ax.set_xlabel('n')
        ax.set_ylabel('Amplitude')

plt.tight_layout()
plt.show()
```
